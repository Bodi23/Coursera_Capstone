```python
#get the data from wikipedia
List_url = "https://en.wikipedia.org/wiki/List_of_postal_codes_of_Canada:_M"
source = requests.get(List_url).text
```

```python
#display the data
soup = BeautifulSoup(source, "html5lib")
table=soup.find('table')
print(soup.title)
from IPython.display import display_html
table1 = str(soup.table)
display_html(tab,raw=True)
```

```python
#import the data into a dataframe
df1 = pd.read_html(table1)
df=df1[0]
df.head()
```

```python
df2 = df[df.Borough != 'Not assigned']
df3 = df2.groupby(['Postal Code','Borough'], sort=False).agg(', '.join)
df3.reset_index(inplace=True)
df3['Neighbourhood'] = np.where(df3['Neighbourhood'] == 'Not assigned',df3['Borough'], df3['Neighbourhood'])
df3
```

```python
#requirement of assingment
df3.shape
```

```python
lat_lon = pd.read_csv('https://cocl.us/Geospatial_data')
lat_lon.head()
```

```python
lat_lon.rename(columns={'Postal Code':'Postal Code'},inplace=True)
df4 = pd.merge(df3,lat_lon,on='Postal Code')
df4.head()
```

```python
df5 = df4[df4['Borough'].str.contains('Toronto',regex=False)]
df5
```

```python
#generate a map of the areas containing Toronto
map_toronto = folium.Map(location=[43.651070,-79.347015],zoom_start=10)

for lat,lng,borough,neighbourhood in zip(df5['Latitude'],df5['Longitude'],df5['Borough'],df5['Neighbourhood']):
    label = '{}, {}'.format(neighbourhood, borough)
    label = folium.Popup(label, parse_html=True)
    folium.CircleMarker(
    [lat,lng],
    radius=3,
    popup=label,
    color='green',
    fill=True,
    fill_color='green',
    fill_opacity=0.7,
    parse_html=False).add_to(map_toronto)
map_toronto
```

```python
#Commence K-Means Clustering using 5 centroids
k=5
Clusters1 = df5.drop(['Postal Code','Borough','Neighbourhood'],1)
TorontoKmeans = KMeans(n_clusters = k,random_state=0).fit(Clusters1)
TorontoKmeans.labels_
df5.insert(0, 'Cluster Labels', TorontoKmeans.labels_)
df5
```

```python
#Recreate the map with the new clusters
ClusterMap = folium.Map(location=[43.651070,-79.347015],zoom_start=10)

#Assign different colours for the clusters
x = np.arange(k)
ys = [i + x + (i*x)**2 for i in range(k)]
MarkerColours = cm.rainbow(np.linspace(0, 1, len(ys)))
SelectedColours = [colors.rgb2hex(i) for i in MarkerColours]

#Add location markers
LocationColours = []
for lat, lon, neighbourhood, cluster in zip(df5['Latitude'], df5['Longitude'], df5['Neighbourhood'], df5['Cluster Labels']):
    label = folium.Popup(' Cluster ' + str(cluster), parse_html=True)
    folium.CircleMarker(
        [lat, lon],
        radius=5,
        popup=label,
        color=SelectedColours[cluster-1],
        fill=True,
        fill_color=SelectedColours[cluster-1],
        fill_opacity=0.7).add_to(ClusterMap)
       
ClusterMap
```
